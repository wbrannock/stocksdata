```python
valued_stocks = stocks.dropna()
valued_stocks.info()
```

```python
valued_stocks.head()
```

```python
companies = pd.read_csv('archive/sp500_companies.csv')
companies.head()
```

```python
df = valued_stocks.merge(companies,
                         how='left',
                         left_on='Symbol',
                         right_on='Symbol',
                         indicator=True)
df.head()
```

```python
df['_merge'].value_counts()
```

```python
df = df[['Date', 'Symbol', 'Industry', 'Adj Close']]
df.head()
```

```python
symbols = set(df['Symbol'].values)
```

```python
df.groupby('Symbol').count()
```

```python
for symbol in symbols:
    df['Symbol'].loc[df['Symbol'] == symbol]
```

```python
df['pctChange'] = df.groupby('Symbol')['Adj Close'].pct_change() * 100
```

```python
df.head()
```

```python
df.loc[df['Symbol'] == 'BBY']
```

```python
df["up_ind"] = (df["pctChange"] > 0).astype(int)
df.head()
```
