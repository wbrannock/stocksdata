```python
valued_stocks = stocks.dropna()
valued_stocks.info()
```

```python
df = valued_stocks[['Date', 'Symbol', 'Industry', 'Adj Close']]
df.head()
```

```python
symbols = set(df['Symbol'].values)
```

```python
df.groupby('Symbol').count()
```

```python
df = df.groupby(['Date', 'Industry']).sum(['Adj Close'])
```

```python
df.reset_index(inplace=True)
```

```python
df
```

```python
df['pctChange'] = df.groupby('Industry')['Adj Close'].pct_change()

df
```

```python
df.loc[df['Industry'] == 'Advertising Agencies']
```

```python
df['up_ind'] = (df['pctChange'] > 0).astype(int)

df
```

```python
df.to_csv('stocks_aggregated.csv', index=False)
```
